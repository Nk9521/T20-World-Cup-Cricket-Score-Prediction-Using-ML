```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
# Convert 'player_dismissed' to numeric: 0 if '0', 1 if not '0'
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == '0' else 1)

# Convert to integer type
df['player_dismissed'] = df['player_dismissed'].astype('int')

# Calculate cumulative sum within each match_id
df['cumulative_dismissals'] = df.groupby('match_id')['player_dismissed'].cumsum()

# Set the number of wickets available per match (assuming 10 wickets)
wickets_per_match = 10

# Calculate wickets_left, ensuring non-negative values
df['wickets_left'] = wickets_per_match - df['cumulative_dismissals']

# Ensure that wickets_left does not drop below 0
df['wickets_left'] = df['wickets_left'].clip(lower=0)

df
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
import pandas as pd

# Convert 'runs_x' column to numeric
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Group by 'match_id'
groups = df.groupby('match_id')

# Initialize list to store rolling sums
last_five = []

# Iterate over each match_id
for match_id in df['match_id'].unique():
    group_df = groups.get_group(match_id)
    
    # Calculate rolling sum with a window size of 30
    rolling_sum = group_df['runs'].rolling(window=30).sum()
    
    # Extend the list with rolling sums (convert to list first to avoid series issues)
    last_five.extend(rolling_sum.fillna(0).tolist())  # Fill NaNs with 0

# Make sure the length of 'last_five' is consistent with DataFrame or handle appropriately
print(last_five)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# Define numerical and categorical columns
numerical_features = ['current_score', 'balls_left', 'wickets_left', 'crr']
categorical_features = ['batting_team', 'bowling_team', 'city', 'last_five']

# Create transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

# Define the pipeline with RandomForestRegressor
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor())
])

# Fit the pipeline
pipe.fit(X_train, y_train)

# Make predictions
y_pred = pipe.predict(X_test)

# Evaluate the model
print("R^2 score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```
