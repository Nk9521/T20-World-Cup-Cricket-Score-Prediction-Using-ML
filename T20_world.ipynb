```python
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm

# Define the directory and collect filenames
directory = '/Users/nandlalkumawat/Desktop/AI_phone/T20word_cup/archive-3/t20s'
filenames = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.yaml')]

# Initialize an empty list to collect dataframes
df_list = []
counter = 1

# Iterate through files and process them
for file in tqdm(filenames):
    try:
        with open(file, 'r') as f:
            data = safe_load(f)
            df = pd.json_normalize(data)
            df['match_id'] = counter
            df_list.append(df)
            counter += 1
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Check if there are any dataframes to concatenate
if df_list:
    final_df = pd.concat(df_list, ignore_index=True)
else:
    final_df = pd.DataFrame()  # Empty DataFrame if no files were processed

# Display the final dataframe
print(final_df)
```

```python
import pandas as pd

count = 1
delivery_df_list = []  # Create a list to hold DataFrames

for index, row in matches.iterrows():
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        count += 1
        continue
    
    count += 1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    
    delivery_df_list.append(loop_df)  # Add the DataFrame to the list

# Concatenate all DataFrames in the list
delivery_df = pd.concat(delivery_df_list, ignore_index=True)
```

```python
delivery_df
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
